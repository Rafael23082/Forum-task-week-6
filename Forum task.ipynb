```python
lloyd = {
 "name": "Lloyd",
 "homework": [90.0,97.0,75.0,92.0],
 "quizzes": [88.0,40.0,94.0],
 "tests": [75.0,90.0]
}
alice = {
 "name": "Alice",
 "homework": [100.0, 92.0, 98.0, 100.0],
 "quizzes": [82.0, 83.0, 91.0],
 "tests": [89.0, 97.0]
}
tyler = {
 "name": "Tyler",
 "homework": [0.0, 87.0, 75.0, 22.0],
 "quizzes": [0.0, 75.0, 78.0],
 "tests": [100.0, 100.0]
}

#students list that prints of the student's name, homework, quizzes and tests
students=[lloyd,alice,tyler]

#Function called average that takes in a list of numbers then returns the average
def average(numbers):
    total=sum(numbers)
    total=float(total)
    num=len(numbers)
    average=total/num
    return average

#Function that asks the user to enter a dictionary (Lloyd, alice or tyler) and then returns the average
def get_average(student):
    hw=(average(student["homework"]))
    quiz=(average(student['quizzes']))
    test=(average(student['tests']))
    score=(hw*0.1)+(quiz*0.3)+(test*0.6)
    return score

#Function that takes in a number then returns the letter grade for the score
def get_letter_grade(score):
    if score>=90:
        return'A'
    elif score>=80:
        return'B'
    elif score>=70:
        return'C'
    elif score>=60:
        return'D'
    else:
        return'F' 

def get_class_average(students):
    results=[]
    for student in students:
        temporary=get_average(student)
        results.append(temporary)
    final=average(results)
    return final

for student in students:
    print(student['name'])
    print(student['homework'])
    print(student['quizzes'])
    print(student['tests'])



class_average=get_class_average(students)
class_letter_grade=get_letter_grade(class_average)
print (f"The score average of the class is {class_average} and the letter grade for the class is {class_letter_grade}")
```

```python
days=("sunday","monday","tuesday","wednesday","thursday","friday","saturday")
total=0
data=[]
maximum=0 #Determine preset value of maximum and minimum
minimum=1000000
for day in days:
    sales=int(input(f"Enter sales for {day}"))
    while sales<0: #To keep requesting user input if the input is less than 0
        sales=int(input(f"Input was invalid. Re-enter the sales for {day}"))
    if sales>maximum:
        maximum=sales
    if sales<minimum: #Replace minimum and maximum value according to the input
        minimum=sales
    print (f"Enter the sales for {day}: {sales}")    
    data.append(sales)
total=sum(data)
print(f"The total sales is: {total}")
print (f"The minimum sales amount is ${minimum}.00 The maximum sale amount is ${maximum}.00")
```
